Feature Detection using Harris Corner Detection      


```
# import libraries
import cv2
import numpy as np
```

```
# Reading the image 
img = cv2.imread('/content/sample_data/object.jpg')
```

```
# Converting the coloured image into Grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img = np.float32(gray_img)
```

```
# Initializing the Harris Corner Detection Algorithm
dst = cv2.cornerHarris(gray_img, blockSize=2, ksize=3, k=0.04)
```

```
# Dilate in order to mark the corners
dst = cv2.dilate(dst, None)
img[dst > 0.01 * dst.max()] = [0, 255, 0]
```

```
# At last, show the final image
from google.colab.patches import cv2_imshow
cv2_imshow(img)
cv2.waitKey()
```
